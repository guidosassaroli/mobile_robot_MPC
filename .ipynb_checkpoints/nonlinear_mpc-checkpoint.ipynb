
# Nonlinear MPC for a Mobile Robot (Unicycle Model)

This notebook implements a **nonlinear MPC** controller for a unicycle-like mobile robot.  
We write the **nonlinear dynamics directly in CasADi** and use direct multiple-shooting with an RK4 integrator.

**Model (continuous time):**
- States: $(x = [p_x, p_y, \theta, v]^\top)$
- Inputs: $(u = [a, \omega]^\top)$  (linear acceleration and yaw rate)
- Dynamics:
  $[
  \dot p_x = v \cos\theta,\quad
  \dot p_y = v \sin\theta,\quad
  \dot \theta = \omega,\quad
  \dot v = a
  ]$

````python
import numpy as np
import casadi as ca
import matplotlib.pyplot as plt
````

````python
plt.rcParams["figure.figsize"] = (6, 4)

def sinusoidal_reference(t, A=2.0, f=0.05, vx=0.8, cx=0.0, cy=0.0):
    """
    2D sinusoidal reference trajectory in position with a forward drift in x.
    Returns [px_ref, py_ref, vx_ref, vy_ref].
    """
    omega = 2.0 * np.pi * f
    px = cx + vx * t
    py = cy + A * np.sin(omega * t)
    vx_ref = vx
    vy_ref = A * omega * np.cos(omega * t)
    return np.array([px, py, vx_ref, vy_ref])
````

### Dynamic nonlinear unicycle model
Continous time.

````python
x_sym = ca.SX.sym('x', 4)  # [px, py, theta, v]
u_sym = ca.SX.sym('u', 2)  # [a, omega]

px, py, th, v = x_sym[0], x_sym[1], x_sym[2], x_sym[3]
a, omg = u_sym[0], u_sym[1]

xdot = ca.vertcat(
    v * ca.cos(th),   # pxdot
    v * ca.sin(th),   # pydot
    omg,              # thdot
    a                 # vdot
)

f_ct = ca.Function('f_ct', [x_sym, u_sym], [xdot])  # continuous-time RHS

# ===== Discretization with RK4 =====
dt = 0.1

def rk4_integrator(f, x, u, dt):
    k1 = f(x, u)
    k2 = f(x + dt/2.0 * k1, u)
    k3 = f(x + dt/2.0 * k2, u)
    k4 = f(x + dt * k3, u)
    return x + dt/6.0 * (k1 + 2*k2 + 2*k3 + k4)

xk = ca.SX.sym('xk', 4)
uk = ca.SX.sym('uk', 2)
xk1 = rk4_integrator(f_ct, xk, uk, dt)
f_d = ca.Function('f_d', [xk, uk], [xk1])  # discrete-time dynamics x_{k+1} = f_d(x_k,u_k)
print("Discrete-time dynamics created.")
````

````python
# ===== MPC setup (nonlinear, direct multiple shooting) =====
N = 25  # horizon

# Weights
Qp = 5.0    # position error weight (x,y)
Qth = 0.0   # heading soft weight (unused here but left for extension)
Qv = 0.2    # speed tracking vs inferred from vx_ref, vy_ref
Ru = 0.05   # input effort
Rd = 0.01   # input rate (for smoothness)
Qf = 10.0   # terminal position weight

# Bounds
a_min, a_max = -2.0, 2.0       # m/s^2
omg_min, omg_max = -2.0, 2.0   # rad/s
v_min, v_max = 0.0, 2.5        # m/s

opti = ca.Opti()

# Decision vars
X = opti.variable(4, N+1)   # states
U = opti.variable(2, N)     # inputs

# Parameters
x0_param = opti.parameter(4)          # current measured state
ref_param = opti.parameter(4, N+1)    # ref over horizon: [px_ref, py_ref, vx_ref, vy_ref]

# Initial condition
opti.subject_to(X[:,0] == x0_param)

# Dynamics constraints
for k in range(N):
    x_next = f_d(X[:,k], U[:,k])
    opti.subject_to(X[:,k+1] == x_next)

# Input bounds and state bounds
opti.subject_to(opti.bounded(a_min, U[0,:], a_max))
opti.subject_to(opti.bounded(omg_min, U[1,:], omg_max))
opti.subject_to(opti.bounded(v_min, X[3,:], v_max))

# Objective: track position, modestly regulate heading & speed, penalize inputs and input rate
obj = 0
for k in range(N):
    px_err = X[0,k] - ref_param[0,k]
    py_err = X[1,k] - ref_param[1,k]
    # desired speed magnitude from (vx_ref, vy_ref); small regularization avoids sqrt(0)
    v_des = ca.sqrt(ref_param[2,k]**2 + ref_param[3,k]**2 + 1e-6)
    v_err = X[3,k] - v_des
    u_k = U[:,k]

    obj += Qp*(px_err**2 + py_err**2) + Qv*(v_err**2) + Ru*ca.sumsqr(u_k)

    if k > 0:
        du = U[:,k] - U[:,k-1]
        obj += Rd*ca.sumsqr(du)

# Terminal position cost
px_errN = X[0,N] - ref_param[0,N]
py_errN = X[1,N] - ref_param[1,N]
obj += Qf*(px_errN**2 + py_errN**2)

opti.minimize(obj)

# Solver
p_opts = {"print_time": False}
s_opts = {"print_level": 0, "warm_start_init_point": "yes", "max_cpu_time": 0.5}
opti.solver('ipopt', p_opts, s_opts)

print("Nonlinear MPC problem constructed.")
````

````python
# ===== Closed-loop simulation =====
T_sim = 30.0
sim_steps = int(T_sim / dt)

# Initial state [px, py, theta, v]
x = np.array([0.0, 0.0, 0.0, 0.5])

nx = 4; nu = 2
xs = np.zeros((nx, sim_steps+1))
us = np.zeros((nu, sim_steps))
refs = np.zeros((4, sim_steps+1))
ts = np.arange(sim_steps+1)*dt

xs[:,0] = x
refs[:,0] = sinusoidal_reference(0.0)

# Warm start
u_guess = np.zeros((nu, N))
x_guess = np.tile(x.reshape(-1,1), (1, N+1))

for t_idx in range(sim_steps):
    t_now = t_idx * dt

    # Build reference over horizon
    ref_h = np.zeros((4, N+1))
    for k in range(N+1):
        ref_h[:,k] = sinusoidal_reference(t_now + k*dt)

    # Set params
    opti.set_value(x0_param, x)
    opti.set_value(ref_param, ref_h)

    # Warm start shift
    if t_idx > 0:
        u_guess = np.hstack([u_guess[:,1:], u_guess[:, -1:]])
        x_guess[:,0] = x
        for k in range(N):
            x_guess[:,k+1] = np.array(f_d(x_guess[:,k], u_guess[:,k]).full()).squeeze()

    opti.set_initial(U, u_guess)
    opti.set_initial(X, x_guess)

    # Solve
    try:
        sol = opti.solve()
    except RuntimeError:
        opti.set_initial(U, 0)
        opti.set_initial(X, 0)
        sol = opti.solve()

    u_opt = sol.value(U)[:,0]

    # Apply first control to the plant
    x = np.array(f_d(x, u_opt).full()).squeeze()

    # Log
    xs[:, t_idx+1] = x
    us[:, t_idx] = u_opt
    refs[:, t_idx+1] = sinusoidal_reference((t_idx+1)*dt)

    # Prepare warm start
    u_guess = sol.value(U)
    x_guess = sol.value(X)

print("Simulation finished.")
````

````python
# ===== Plots =====

# XY trajectory
plt.figure()
plt.plot(refs[0,:], refs[1,:], label="reference (pos)")
plt.plot(xs[0,:], xs[1,:], label="robot (pos)")
plt.xlabel("x [m]")
plt.ylabel("y [m]")
plt.title("XY Trajectory (Nonlinear MPC)")
plt.axis("equal")
plt.legend()
plt.grid(True)
plt.show()

# Position vs time
plt.figure()
plt.plot(ts, refs[0,:], label="x_ref")
plt.plot(ts, xs[0,:], label="x")
plt.plot(ts, refs[1,:], label="y_ref")
plt.plot(ts, xs[1,:], label="y")
plt.xlabel("time [s]")
plt.ylabel("position [m]")
plt.title("Position Tracking")
plt.legend()
plt.grid(True)
plt.show()

# Heading and speed
plt.figure()
plt.plot(ts, xs[2,:], label="theta")
plt.plot(ts, xs[3,:], label="v")
plt.xlabel("time [s]")
plt.ylabel("state")
plt.title("Heading and Speed")
plt.legend()
plt.grid(True)
plt.show()

# Inputs
plt.figure()
plt.step(ts[:-1], us[0,:], where="post", label="a (m/s^2)")
plt.step(ts[:-1], us[1,:], where="post", label="omega (rad/s)")
plt.xlabel("time [s]")
plt.ylabel("input")
plt.title("Control Inputs")
plt.legend()
plt.grid(True)
plt.show()
````


## Ideas to Extend

- Add obstacles as soft constraints (e.g., barrier cost on distance to obstacles).
- Track a full pose reference by adding a heading cost \(Q_\theta\) and defining a desired \(\theta_d\).
- Enforce terminal ingredients (terminal set or LQR terminal cost).
- Try a spline/waypoint reference instead of an analytic one.
